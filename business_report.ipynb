```python
import numpy as np
import pandas as pd
import datetime
```

```python
df=pd.read_csv("business_report.csv")
df.head()
```

```python
df_new=df[df.columns[0:3]]
```

```python
df_new.head()
```

```python
def money_to_float(money_str):
    
    
    return float(money_str.replace("$",""))
```

```python
df_new1=df_new.copy()
df_new1["Ordered Product Sales"]=df_new1["Ordered Product Sales"].apply(money_to_float)
df_new1["Ordered Product Sales - B2B"]=df_new1["Ordered Product Sales - B2B"].apply(money_to_float)
```

```python
df_new1.head()
```

```python
df_new1["Total Product Sales"]=df_new1["Ordered Product Sales"] + df_new1["Ordered Product Sales - B2B"]
df_new1.head()
```

```python
def convert_date_format(date_str):
    
    return datetime.datetime.strptime(date_str,"%m/%d/%Y").strftime("%Y-%m-%d")
```

```python
df_new1["New_Date_format"]=df_new1["Date"].apply(convert_date_format)
df_new1.head()
```

```python
df_new1["Total_Sales_Column"]="Total Product Sales"
df_new1.head()
```

```python
df_final=df_new1[df_new1["Total Product Sales"]>0]
df_final.head()
```

```python
column_order=["Total_Sales_Column","New_Date_format","Total Product Sales"]
df_final=df_final[column_order]
df_final.head()
```
